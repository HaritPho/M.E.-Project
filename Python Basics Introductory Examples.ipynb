Plotting Interest Rates

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```

```python
T = 200
alpha = 0.9
x_array = np.empty(T+1)
x_array[0] = 0

for value in range(0,T):
    x_array[value+1] = alpha * x_array[value] + np.random.randn()
    
plt.plot(x_array)
plt.show()
```
